```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Tree Map Total Employee Per Context
data = pd.read_csv("Messy_HR_Dataset_Detailed.csv")

context_col = ['Title','BusinessUnit','Division','State']

for dt in context_col:
    total_context = data.groupby(dt)['Employee ID'].count().reset_index()
    
    # Sort 
    total_context = total_context.sort_values(by='Employee ID', ascending=False)
    
    # Extract
    sizes = total_context['Employee ID']
    labels = total_context.apply(lambda x: f"{x[dt]}\n({x['Employee ID']})", axis=1)
    
    # Plot Tree Map
    plt.figure(figsize=(15, 9))
    squarify.plot(sizes=sizes, label=labels, alpha=0.8)
    plt.title("Number of Employee", fontsize=14)
    plt.axis('off')
    plt.show()
```
