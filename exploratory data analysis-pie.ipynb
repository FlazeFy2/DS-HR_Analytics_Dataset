```python
import pandas as pd
import matplotlib.pyplot as plt
import re

df = pd.read_csv("Messy_HR_Dataset_Detailed.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = [
    'Title', 'Supervisor', 'BusinessUnit', 'EmployeeStatus', 'EmployeeType', 'PayZone', 'EmployeeClassificationType',
    'DepartmentType', 'TerminationType', 'Division', 'State', 'GenderCode', 'JobFunctionDescription', 'RaceDesc',
    'MaritalDesc', 'Performance Score', 'Training Program Name', 'Training Type', 'Training Outcome', 'Location', 'Trainer'
]

charts = [] 

for dt in target_col:
    # Check Unique Value
    unique_count = df[dt].nunique()
    top_context = df[dt].value_counts().head(5)

    # Clean col name for title
    formatted_name = re.sub(r'(?<!^)(?=[A-Z])', ' ', dt)

    # Title Condition
    if unique_count > 5:
        title = f"Top 5 {formatted_name}"
    else:
        title = f"{formatted_name} Comparison"

    charts.append({
        "title": title,
        "data": top_context
    })

# Pie Chart
for chart in charts:
    plt.figure(figsize=(8, 8))
    chart["data"].plot.pie(autopct="%1.1f%%", startangle=90)
    plt.title(chart["title"])
    plt.ylabel("")
    plt.show()
```
